```python
#Impoto librerias
import numpy as np #para el manejo de arrays
from keras.models import Sequential
from keras.layers.core import Dense

#cargo las 4 combinaciones de las compuertas XOR
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

#los resultados que se obtienen en el mismo orden
target_data = np.array([[0],[1],[1],[0]], "float32")
```

```python
#Creo arquitectura de la red neuronal
model = Sequential() #creo una serie de capas de neuronas secuenciales, es decir una delante de otra

#input_dim=2 creo la capa de entrada con dos neuronas, 16 neuronas en la capa oculta y la funcion activacion relu
model.add(Dense(16, input_dim=2, activation='relu'))

#creo una capa de salida de una neurona y funcion activacion sigmoid
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```python
#Entreno la red
model.fit(training_data, target_data, epochs=1000) 

#indico con model.fit() las entradas y sus salidas y la cantidad de iteraciones de aprendizaje (epochs) de entrenamiento
```

```python
#Evalo el modelo
scores = model.evaluate(training_data, target_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
#Realizo las 4 predicciones posibles XOR pasando las entradas
print (model.predict(training_data).round())
```
